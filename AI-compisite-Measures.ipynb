```python
 # Insert a new column and add the indicators

'''For coursera indicators'''

coursera1 = pd.read_csv("23. Coursera AI Skill Percentile Rank - 2018 (skill percentile_rank).csv")
coursera1['DocType'] = "skill_percentile_rank"

coursera2 = pd.read_csv("22. Coursera_ AI Enrollment Share (enroll_share).csv")
coursera2['DocType'] = "enroll_share"

''' For Linkedin indicators'''

# linkedin2 =  pd.read_csv("linkedIn_AI_hires_count.csv")
# linkedin2['DocType'] = "Total number of hires on LinkedIn"

linkedin =  pd.read_csv("26.  LinkedIn_ AI Hiring Index (ai_hiring_index).csv")
linkedin['DocType'] = "ai_hiring_index"

# linkedin3 =  pd.read_csv("linkedIn_AI_member_count.csv")
# linkedin3['DocType'] = "Number of LinkedIn members with AI skills"

'''For nesta indicators '''

nesta1 = pd.read_csv("19. NESTA - arXiv_ Number of Deep Learning papers on arXiv (num_DLpapers).csv")
nesta1['DocType']="num_DLpapers"

nesta2 = pd.read_csv("21. NESTA - arXiv Deep Learning Papers RCA (rca_DLpapers).csv")
nesta2['DocType']="rca_DLpapers"

# nesta3 = pd.read_csv("32. NESTA arXiv_ Female AI authors (female_ai_authors).csv")
# nesta3['DocType']="female_ai_authors"

'''For Quid indicators '''

quid1 = pd.read_csv("27. CAPIQ, CB, and Quid_ AI Private Investment (ai_privateinvestment_tot).csv")
quid1['DocType']="ai_privateinvestment_tot"

quid2 = pd.read_csv("29. CAPIQ, CB, Quid_ Number of AI Startups Funded (num_AIstartups).csv")
quid2['DocType']="num_AIstartups"

'''Robot Installations'''

robot = pd.read_csv("31. IFR and World Robotics_ Robot Installations (robot).csv")
robot['DocType']="robot"
```

```python
# creating the list of dataframes to concat
finaldata = [coursera1,coursera2,linkedin,nesta1,nesta2,quid1,quid2,robot]
# concat to a single final dataset
finaldata1 = pd.concat(finaldata)
# dropping unnecessary columns
#finaldata1 = finaldata1.drop(['2008','2009','2010','2011','2012','2013','2014','2019'],axis=1)
#Rearranging the columns
finaldata1 = finaldata1[['country','DocType','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']]
finaldata1.to_csv("D:\\composite_measures\\OutputFiles\\8IndicatorsData.csv",index = False)
```

```python
# reshaping the dataframe using  Melt method.
finaldata1 = finaldata1.melt(id_vars=["country", "DocType"], var_name="PublishYear", value_name="Count")

# converting the year column to numneric as it is in string datatype
finaldata1["PublishYear"] = pd.to_numeric(finaldata1["PublishYear"])

# cross verifying and removing the values lesser than 2014 
#finaldata1 = finaldata1[finaldata1['PublishYear']>=2015]
```

```python
'''For Linkedin & NESTA Indicators'''

LinkedIn_1 = pd.read_csv("24, 25. LinkedIn_ AI Skill Penetration and number of AI occupations (skill_penetration, num_AIocc).csv")

LinkedIn_2 = pd.read_csv("33, 35. LinkedIn_ Female AI Skill Penetration and Female Number of AI occupations (female_ai_skill_penetration, num_AIocc_fem).csv")

LinkedIn_3 = pd.read_csv("34, 36. LinkedIn_ Male AI Skill Penetration and Male number of AI occupations (male_ai_skill_penetration, num_AIocc_male).csv")


nesta = pd.read_csv("32. NESTA arXiv_ Female AI authors (female_ai_authors).csv")

final = [LinkedIn_1,LinkedIn_2,LinkedIn_3,nesta]
final1 = pd.concat(final)
final1.fillna(0)
final1 = final1[['country','year','skill_penetration','num_AIocc','female_ai_skill_penetration','num_AIocc_fem','male_ai_skill_penetration','num_AIocc_male','female_ai_authors']]
final1.to_csv("D:\\composite_measures\\OutputFiles\\7IndicatorsData.csv",index = False)
```

```python
'''MAG Data'''
 # Insert a new column and add the indicator

mag1 = pd.read_csv("MAG References_ AI Conference paper, journal, and patent citations (num_AIconf_references, num_AIjournal_references, num_AIpatent_references).csv")
# drop the unnecessary column
mag1 = mag1.drop(['ReferenceCountry'],axis=1)
mag1 = mag1.groupby(['Country','PublishYear','DocType']).sum().reset_index()
mag1['DocType']= mag1['DocType'].str.replace("Journal", "num_AIjournal_references", case = False) 
mag1['DocType']= mag1['DocType'].str.replace("Conference", "num_AIconf_references", case = False) 
mag1['DocType']= mag1['DocType'].str.replace("Patent", "num_AIpatent_references", case = False)

mag2 = pd.read_csv("MAG Citations_ AI Conference paper, journal, and patent citations (num_AIconf_citation, num_AIjournal_citation, num_AIpatent_citation).csv")
mag2 = mag2.drop(['CitationCountry'],axis=1)
mag2 = mag2.groupby(['Country','PublishYear','DocType']).sum().reset_index()
mag2['DocType']= mag2['DocType'].str.replace("Journal", "num_AIjournal_citation", case = False) 
mag2['DocType']= mag2['DocType'].str.replace("Conference", "num_AIconf_citation", case = False) 
mag2['DocType']= mag2['DocType'].str.replace("Patent", "num_AIpatent_citation", case = False)

mag3 = pd.read_csv("MAG Volume_ AI Journals, Conference Publications and Patents (num_AIconf_papers, num_AIjournal_papers, num_AIpatents).csv")
mag3.rename(columns ={'NumPaper' : 'Count'}, inplace = True)
mag3 = mag3.groupby(['Country','PublishYear','DocType']).sum().reset_index()
mag3['DocType']= mag3['DocType'].str.replace("Journal", "num_AIjournal_papers", case = False) 
mag3['DocType']= mag3['DocType'].str.replace("Conference", "num_AIconf_papers", case = False) 
mag3['DocType']= mag3['DocType'].str.replace("Patent", "num_AIpatents", case = False) 


# Merge and concat
mag = [mag1,mag2,mag3]
magfinal = pd.concat(mag)

magfinal.rename(columns ={'Country' : 'country'}, inplace = True)

# The other 9 indicators above have the values starting from 2014, hence considering the same for MAG to avoid messy data
#magfinal = magfinal[magfinal['PublishYear']>=2015]

# save the dataframe
magfinal.to_csv("D:\\composite_measures\\OutputFiles\\9IndicatorsMagData.csv", index= False)
```
